```python
# Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, roc_curve, auc,
                             classification_report)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

import joblib
import io
import zipfile

# Step 2: File Upload (Google Colab)
try:
    from google.colab import files
    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]

    # If uploaded file is ZIP, extract CSV
    if file_name.endswith(".zip"):
        with zipfile.ZipFile(io.BytesIO(uploaded[file_name]), 'r') as z:
            print("Files in ZIP:", z.namelist())
            csv_files = [f for f in z.namelist() if f.endswith('.csv')]
            if len(csv_files) == 0:
                raise ValueError("No CSV file found in ZIP!")
            csv_file = csv_files[0]
            with z.open(csv_file) as f:
                try:
                    df = pd.read_csv(f, encoding='utf-8')
                except UnicodeDecodeError:
                    df = pd.read_csv(f, encoding='latin1')
    else:
        try:
            df = pd.read_csv(io.BytesIO(uploaded[file_name]), encoding='utf-8')
        except UnicodeDecodeError:
            df = pd.read_csv(io.BytesIO(uploaded[file_name]), encoding='latin1')

except ImportError:
    # For Local Jupyter Notebook
    print("Google Colab not detected. Please provide a local path to CSV or ZIP:")
    file_path = input("Enter file path: ")

    if file_path.endswith(".zip"):
        with zipfile.ZipFile(file_path, 'r') as z:
            print("Files in ZIP:", z.namelist())
            csv_files = [f for f in z.namelist() if f.endswith('.csv')]
            if len(csv_files) == 0:
                raise ValueError("No CSV file found in ZIP!")
            csv_file = csv_files[0]
            with z.open(csv_file) as f:
                try:
                    df = pd.read_csv(f, encoding='utf-8')
                except UnicodeDecodeError:
                    df = pd.read_csv(f, encoding='latin1')
    else:
        try:
            df = pd.read_csv(file_path, encoding='utf-8')
        except UnicodeDecodeError:
            df = pd.read_csv(file_path, encoding='latin1')

print("Dataset Loaded Successfully!")
print(df.head())

# Step 3: Data Info
print("\nDataset Info:")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())

# Step 4: Handle Missing Values / Duplicates
df = df.drop_duplicates()
df = df.dropna()

# Step 5: Encode Categorical Columns
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category').cat.codes

print("\nData After Encoding:\n", df.head())

# Step 6: Basic Stats & EDA
print("\nStatistical Summary:\n", df.describe())

# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Interactive Plot (if age, chol, target exist) - FIXED THIS LINE
if 'age' in df.columns and 'chol' in df.columns and 'target' in df.columns:
    fig = px.scatter(df, x="age", y="chol", color="target", title="Age vs Cholesterol by Target")
    fig.show()
```
